### 네거티브 포지티브 반전하기

NumPy를 사용한 필터 처리의 예로 이미지를 네거티브-포지티브 반전하는 프로그램을 만들어 봅시다.

네거티브 [negative]란?
피사체와 흑백 부분의 관계가 반대로 되어 있는 사진 필름이나 화상(畵像). 간단히 네거라고 하며 우리말로는 음화(陰畵)라고 한다. 

카메라로 촬영한 필름 등의 감광판을 현상액에 담그면 잠상(潛像)부분만이 검게 되어 피사체의 흑백 부분과는 반대의 화상이 얻어지는데, 이것을 다시 정착액에 넣어 더 이상 현상이 진행되지 못하도록 안정시킨 뒤에 물로 씻어 건조시키면 네거티브가 얻어진다. 

이러한 네거티브와는 달리 피사체와 흑백 부분의 관계가 같게 되어 있는 사진 필름이나 화상은 포지티브(positive) 또는 간단히 포지라고 하며 우리말로는 양화(陽畵)라고 한다.

포지티브 [positive]란?
음화. 네거티브의 반대. 줄여서 포지(posi)라 하고, 이와 같은 경우의 필름을 포지 필름이라 한다. 

㉠포지티브한 영상을 만들어내도록 고안된 미사용 필름. 

㉡피사체의 색조와 똑같은 색조를 내는 현상한 필름.

프로그램을 실행하면 다음과 같이 출력합니다. 

네거티브 - 포지티브 반전 처리는 'img = 255 - img' 라는 한 줄의 처리로 이루어집니다. 

각 픽셀은 RGB로 표현되며 각 요소는 0 부터 255까지의 수치로 나타냅니다. 

NumPy 를 사용하면 모든 요소에 대한 처리를 이렇게 쉽게 할 수 있습니다.

```python
import matplotlib.pyplot as plt
import cv2

# 이미지 읽어 들이기
img = cv2.imread("girl.jpg")

# 네거티브 반전
img = 255 - img

# 이미지 출력하기
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

### 그레이스케일로 변환하기

아울러 머신러닝을 할 때는 색정보를 낮춰 그레이스케일로 처리하는 경우가 많습니다. 

이미지를 그레이스케일로 변환할 때는 cv2.cvtColor() 함수를 사용합니다. 

다음 프로그램은 색상 이미지를 그레이스케일로변환해 출력합니다.

```python
import matplotlib.pyplot as plt
import cv2

# 이미지 읽어 들이기
img = cv2.imread("girl.jpg")

# 색공간을 그레이스케일로 변환하기
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 이미지 출력하기
plt.imshow(img, cmap="gray")
plt.axis("off")
plt.show()
```

색공간 변환을 하는 cv2.cvtColor() 함수의 2 번째 매개변수에는 다음과 같은 매개변수를 지정할 수 있습니다. 

참고로 OpenCV는 150종류 이상의 색공간 변환을 지원합니다. 

다음은 자주 사용되는 것을 표로 정리한 것입니다.


| 상수 | 결과 |
|-|-|
| cv2.COLOR_BGR2GRAY | BGR 색상 이미지를 그레이스케일로 변환 |
| cv2.COLOR_RGB2BGR | RGB 색상 이미지를 BGR 색상 이미지로 변환 |
| cv2.COLOR_BGR2YCrCb | BGR 색상 이미지를 YCrCb로 변환 |
| cv2.COLOR_BGR2HSV | BGR 색상 이미지를 HSV로 변환 |

다음은 모든 색상의 상수를 출력한 것입니다.

```python
import cv2
[i for i in dir(cv2) if i.startswith('COLOR_')]
```

```python
import matplotlib.pyplot as plt
import cv2

# 이미지 읽어 들이기
img = cv2.imread("girl.jpg")
# 색공간을 그레이스케일로 변환하기
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 역치 지정하기
th = 90
img[img > th] = 255
img[img < th] = 0

# 이미지 출력하기
plt.imshow(img, cmap="gray")
plt.show()
```

```python
# 이미지 읽어 들이기
img = cv2.imread("girl.jpg")

# 이미지의 엣지 검출
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img = cv2.Canny(img, 100, 200, 255)
img = 255 - img

# 이미지 저장하기
cv2.imwrite("out.png", img)

# 이미지 출력하기
#plt.axis("off")
plt.imshow(img, cmap="gray")

cv2.imwrite('toon.png', img)
plt.show()
```

### 이미지 자르기

이어서 이미지의 얼굴 부분을 자르고 크기를 변경해 보겠습니다. 

이미지를 자를 때는 리스트 슬라이스를 사용해 

'배열[y1:y2,x1:x2]' 형식으로 이미지를 지를 수 있습니다.

```python
import matplotlib.pyplot as plt
import cv2

# 이미지 읽어 들이기
img = cv2.imread("girl.jpg")

# 이미지 자르기
im2 = img[300:700, 350:700]

# 이미지 크기 변경하기
im2 = cv2.resize(im2, (400, 400))

# 크기 변경한 이미지 저장하기
cv2.imwrite("cut-resize.png", im2)

# 이미지 출력하기
plt.imshow(cv2.cvtColor(im2, cv2.COLOR_BGR2RGB))
plt.show()
```

#### 좌우 반전과 상하반전

'cv2.flip()' 을 사용하면 이미지를 좌우 반전이나 상하 반전할 수 있습니다.

cv2.flip() 함수는 다음과 같은 형태로 사용합니다.

[서식] 이미지 반전
cv2.flip(이미지 데이터 / 반전 방향)

반전 방항을 0으로 지정하면 상하 반전, 1 로 지정하면 좌우 반전이 됩니다.

다음 이미지는 반전 방항을 0으로 지정한 경우입니다.

```python
import matplotlib.pyplot as plt
import cv2

# 이미지 읽어 들이기
img = cv2.imread("girl.jpg")

# 원래 이미지를 왼쪽에 출력하기
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

# 이미지 좌우 반전하기
plt.subplot(1, 2, 2)
img2 = cv2.flip(img, 0)
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.show()
```

앞에서는 얼굴을 검출할 때 scipy의 ndimage 모듈을 사용하는 방법을 소개했습니다. 

OpenCV로 이미지 회전을 하는 경우는 다음과 같이 합니다.

```python
import matplotlib.pyplot as plt
import cv2

# 이미지 읽어 들이기
img = cv2.imread("girl.jpg")

# 이미지 크기 추출하기 ---- (1)
h, w, colors = img.shape
size = (w, h)

# 이미지의 중심 위치 추출하기 ---- (2)
center = (w // 2, h // 2)

# 회전 변환 행렬 만들기 ---- (3)
angle = 45
scale = 1.0
matrix = cv2.getRotationMatrix2D(center, angle, scale)

# 변환하기 ---- (4)
img2 = cv2.warpAffine(img, matrix, size)

# 원래 이미지를 왼쪽에 출력하기
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

# 회전한 이미지를 오른쪽에 출력하기
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.show()

# 프로그램을 실행하면 다음과 같이 반시계 방향으로 45도 회전된 이미지를 출력합니다.
```

프로그램을 살펴봅시다.

프로그램 (※ 1) 부분에서는 이미지 크기를 추출합니다. 

imread()로 읽어 들인 numpy.ndarray 자료형의 데이터는 shape 속성을 사용해 튜플 형태로 (이미지의 높이, 이미지의 너비, 색 깊이)를 확인할 수 있습니다. 

이미지의 너비와 높이를 2로 나누면 중심 위치를 알 수 있습니다. 

프로그램 (※ 3) 부분에서는 getRotation Matrix2D() 함수를 사용해 2차원 회전 변환 행렬을 생성하고 

프로그램 (※ 4) 부분에서는 warpAffine() 함수를 사용해 실제로 회전 처리를 합니다.

이때 getRotationMatrix2D()는 다음과 같은 형식으로 회전 변환 행렬을 추출합니다.

[서식] 회전 변환 행렬 추출하기
matrix = cv2.getRotationMatrix2D(중심점 / 회전각도 / 확대비율)

중심점은 (x. y) 형태의 튜플로 지정하고 각도는 0도부터 360도까지 지정합니다. 

확대율은 1.0과 같은 배수를 지정하며 1O보다 큰 값을 지정하면 확대됩니다.
